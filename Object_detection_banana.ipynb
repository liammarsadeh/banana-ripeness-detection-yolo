```python
from ultralytics import YOLO
```

## Call YOLOv8n Model
- our dataset is quite small and it's for expermintation so we don't need a large YOLO model like `YOLOv8s with 4x param, or yolov8m with 8x param` latency of model will be increase and it's not effeciency to make our model complex when we can have simpler one

```python
model = YOLO('yolov8n.yaml')
```

```python
model.train(data='banana dataset/data.yaml',epochs=35,imgsz=640,device=0)
```

```python
model.val(data='banana dataset/data.yaml',split='test')
```

## Conclusion of first Train model
- I have trained model for 5 epochs for expermintaion purposes i had to see if `mps, Loss` will have chance to improve it shows that mps is increasing in both `Val, Train` and Loss is decreasing in both `Val, Train` that gives a generalization expectation
- After that model has been expermntize by testing data and actually it's a good start
